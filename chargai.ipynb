# **Numpy**

```python
# Sample dataset: 5 samples, 3 features (like in a real ML task)
data = np.array([
    [5.1, 3.5, 1.4],
    [4.9, 3.0, 1.4],
    [6.2, 3.4, 5.4],
    [5.9, 3.0, 5.1],
    [5.5, 2.3, 4.0]
])

print("Original Data:\n", data)
```

```python
# 1. Array creation
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2], [3, 4]])
print("Array 1:", arr1)
print("Array 2:\n", arr2)
# arr3=df[""]

```

```python
# 2. Basic arithmetic operations
sum_arr = arr1 + 10
mul_arr = arr1 * 2
print("\nArray 1 + 10:", sum_arr)
print("Array 1 * 2:", mul_arr)
```

```python
# 3. Indexing and slicing
print("\nFirst element of Array 1:", arr1[0])
print("Slice of Array 1 (elements 2 to 4):", arr1[1:4])
print("Element at row 0, column 1 of Array 2:", arr2[0, 1])
```

```python
# 1. Basic statistics
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
print("\nMean of each feature:", mean)
print("Standard Deviation of each feature:", std)
```

```python
# 2. Normalization (Standard Score / Z-score normalization)
normalized_data = (data - mean) / std
print("\nNormalized Data:\n", normalized_data)
```

```python
# 4. Element-wise operations
squared_data = np.square(data)
print("\nElement-wise Squared Data:\n", squared_data)
```

```python
# 5. Reshaping (useful in data preprocessing)
flattened = data.flatten()
reshaped = flattened.reshape(5, 3)
print("\Flattened Data:\n", flattened)
print("\nReshaped Data:\n", reshaped)
```

# **Matplotlib**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 1. Line Plot
plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sin(x)', linestyle='--')
plt.plot(x, y2, label='cos(x)')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate data for histogram
data_hist = np.random.randn(1000)

# 4. Histogram
plt.figure(figsize=(8, 6))
plt.hist(data_hist, bins=30, color='red', edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Generate data for scatter plot
np.random.seed(20)
x_scatter = np.random.rand(50) * 10
y_scatter = 2 * x_scatter + 1 + np.random.randn(50) * 2

# 2. Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='red', marker='+')
plt.title('Scatter Plot of Random Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Generate data for bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 39]

# 3. Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='green')
plt.title('Bar Plot of Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y')
plt.show()
```
